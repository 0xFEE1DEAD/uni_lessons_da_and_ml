````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Построение линейного графика
plt.plot(np.arange(1, 6), np.arange(1, 6) * 10)
# Добавить заголовок
plt.title("Пример линейного графика")
# Разметить x и y
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
# добавить сетку
plt.grid(alpha=0.3)
plt.show()
````

````python
# Изменить размер фигуры
plt.figure(figsize=(2, 4))
plt.plot(np.arange(1, 6), 10 * np.arange(1, 6))
plt.show()
````

````python
# Несколько графиков
multi_graph_x = np.arange(1, 6)
multi_graph_y1 = np.arange(1, 6) * 10
multi_graph_y2 = np.arange(1, 6)[::-1] * 10

plt.plot(multi_graph_x, multi_graph_y1, color="red")
plt.plot(multi_graph_x, multi_graph_y2, color="green")
plt.grid()
plt.show()
````

````python
# точечный график (диаграмма рассеяния)
from numpy.random import default_rng

x = np.floor(10 * default_rng(52).random(20))
y = np.floor(10 * default_rng(42).random(20))

plt.scatter(x, y)
plt.show()
````

````python
# Круговая диаграмма
x = [2, 3, 1.5]
plt.pie(x, labels=["1", "2", "3"])
plt.show()
````

````python
# Столбчатая диаграмма
x = ["1", "2", "3", "4"]
heights = [3, 4, 5, 2]
plt.bar(x, heights)
plt.show()
````

````python
# Построение гистограммы
values = default_rng(42).random(100)
plt.hist(values)
plt.show()
````

````python
# построение нескольких графиков
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].pie([1, 3, 7])
ax[1].bar(["Cat 1", "Cat 2"], [1, 5])
plt.show()
````
